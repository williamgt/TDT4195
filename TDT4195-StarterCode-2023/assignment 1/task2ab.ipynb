```python
import matplotlib.pyplot as plt
import pathlib
from utils import read_im, save_im
output_dir = pathlib.Path("image_solutions")
output_dir.mkdir(exist_ok=True)
```

```python
im = read_im(pathlib.Path("images", "duck.jpeg"))
plt.imshow(im)
```

## Task 2a

```python
def greyscale(im):
    """ Converts an RGB image to greyscale
    
    Args:
        im ([type]): [np.array of shape [H, W, 3]]
    
    Returns:
        im ([type]): [np.array of shape [H, W]]
    """

    def convert_rg_grayscale(r,g,b):
        return 0.212*r + 0.7152*g + 0.0722*b

    for i in range(im.shape[0]):
        for j in range(im.shape[1]):
            im[i][j] = convert_rg_grayscale(im[i][j][0],im[i][j][1],im[i][j][2]) #Loop through every pixel and convert them to grayscale


    return im
```

```python
im_greyscale = greyscale(im)
save_im(output_dir.joinpath("duck_greyscale.jpeg"), im_greyscale, cmap="gray")
plt.imshow(im_greyscale, cmap="gray")
```

## Task 2b

```python
def inverse(im):
    """ Finds the inverse of the greyscale image
    
    Args:
        im ([type]): [np.array of shape [H, W]]
    
    Returns:
        im ([type]): [np.array of shape [H, W]]
    """

    def inverse_transform(p):
        return 1 - p

    for i in range(im.shape[0]):
        for j in range(im.shape[1]):
            im[i][j] = inverse_transform(im[i][j][0]) #Loop through every pixel and invert it assuming already grayscale

    return im
```

```python
im_grayscale_inverse = inverse(im_greyscale)
save_im(output_dir.joinpath("duck_greyscale_inverse.jpeg"), im_grayscale_inverse, cmap="gray")
plt.imshow(im_grayscale_inverse, cmap="gray")
```
